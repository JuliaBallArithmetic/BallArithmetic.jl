```julia
import Pkg; Pkg.activate("./")
Pkg.add("Plots")
```

```julia
A = rand(10, 10)
```

```julia
using LinearAlgebra
F  =schur(Complex.(A))
```

```julia
T = F.T
```

```julia
eigs = diag(F.T)
rho = 0.1

circ = [eigs[1]+rho*exp(2*π*θ*im) for θ in 0:0.01:1]
```

```julia
vals = [svd(z*I-T).S for z in circ]
```

```julia
using Plots
```

```julia
plt = plot();
```

```julia
for i in 1:10
    plot!(plt, 0:0.01:1, [x[i] for x in vals], label = "")
end
plt
```
